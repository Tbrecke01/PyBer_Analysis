### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
print(pyber_data_df.head())
print("Total rows: " + str(pyber_data_df['ride_id'].count()))
```

## Deliverable 1: Get a Summary DataFrame 

```python
#Create DF for City type
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
```

```python
urban_cities_df.head()
```

```python
#  1. Get the total rides for each city type
#Urban Cities
urban_ride_count = urban_cities_df["ride_id"].count()
print(urban_ride_count)

#Subrban Cities
suburban_ride_count = suburban_cities_df["ride_id"].count()
print(suburban_ride_count)

#Rural Cities
rural_ride_count = rural_cities_df["ride_id"].count()
print(rural_ride_count)

# print(pyber_data_df.groupby('type')["ride_id"].count())
```

```python
# 2. Get the total drivers for each city type
#Urban
driver_count = city_data_df.groupby(["type"])["driver_count"].sum()
print(driver_count)
```

```python
#  3. Get the total amount of fares for each city type
#Urban
urban_fare = urban_cities_df.groupby(["type"])["fare"].sum()
print(urban_fare)

#Suburban
suburban_fare = suburban_cities_df.groupby(["type"])["fare"].sum()
print(suburban_fare)

#Rural
rural_fare = rural_cities_df.groupby(["type"])["fare"].sum()
print(rural_fare)
```

```python
#Alt Totam Amount of Fares
print(pyber_data_df.groupby('type')["fare"].sum())
```

```python
#  4. Get the average fare per ride for each city type. 
#urban
urban_ride_fare = urban_fare / urban_ride_count
print(urban_ride_fare)

#Suburban
suburban_ride_fare = suburban_fare / suburban_ride_count
print(suburban_ride_fare)

#Rural
rural_ride_fare = rural_fare / rural_ride_count
print(rural_ride_fare)
```

```python
#Alt fare per ride
ride_fare = (pyber_data_df.groupby('type')["fare"].sum()) / (pyber_data_df.groupby('type')["ride_id"].count())
print(ride_fare)
```

```python
# 5. Get the average fare per driver for each city type. 
#urban
urban_drive_fare = urban_fare / driver_count["Urban"]
print(urban_drive_fare)

#Suburban
suburban_drive_fare = suburban_fare / driver_count["Suburban"]
print(suburban_drive_fare)

#Rural
rural_drive_fare = rural_fare / driver_count["Rural"]
print(rural_drive_fare)
```

```python
#Alt fare per driver
driver_fare = (pyber_data_df.groupby('type')["fare"].sum()) / driver_count
print(driver_fare)
```

```python
pyber_summary_df = pd.DataFrame({
    "Total Rides": pyber_data_df.groupby('type')["ride_id"].count(),
    "Total Drivers": driver_count,
    "Total Fares": (pyber_data_df.groupby('type')["fare"].sum()),
    "Average Fare per Rider": ride_fare,
    "Average Fare per Driver": driver_fare
})


pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
graph_df = pyber_data_df.groupby(["type","date"])["fare"].sum()
graph_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
graph_df = graph_df.reset_index()
graph_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
graph_pivot = graph_df.pivot(index = "date", columns = "type", values = 'fare')
graph_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pivot_df = graph_pivot.loc["2019-01-01":"2019-04-29"]
pivot_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
pivot_df.index = pd.to_datetime(pivot_df.index)
pivot_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pivot_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
week_fares_df = pivot_df.resample("W").sum()
week_fares_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
week_fares_df.plot(figsize = (15,5))

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
plt.title('Total Fare by City Type')
plt.xlabel(' ')
plt.ylabel('Fare ($USD)')

plt.savefig("resources/PyBer_fare_summary.png")
```
